```python
import pandas as pd
df3=pd.read_csv("Products.csv")
df3.index.name='s_no'

# Step 2: Remove unwanted characters (dollar signs, periods, spaces, commas)
df3['Unit Cost USD'] = df3['Unit Cost USD'].replace({r'\$': '', r'\.': '', r',': '', r'\s+': ''}, regex=True)
df3['Unit Price USD'] = df3['Unit Price USD'].replace({r'\$': '', r'\.': '', r',': '', r'\s+': ''}, regex=True)

# Step 3: Convert to numeric (float), setting errors='coerce' to handle invalid data gracefully
df3['Unit Cost USD'] = pd.to_numeric(df3['Unit Cost USD'], errors='coerce')
df3['Unit Price USD'] = pd.to_numeric(df3['Unit Price USD'], errors='coerce')

Products_df=df3
```
